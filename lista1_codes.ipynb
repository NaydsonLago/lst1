```python
#questao 3 - composição das duas imagens do item 1
from PIL import Image
import cv2
import matplotlib.pyplot as plt
import numpy as np

image1 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Images/img1.jpg')
image2 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Images/img2.png')
#as duas imagens precisam possuir o mesmo tamanho
dim = (1200, 800) #novas dimensões
resized_image1 = cv2.resize(image1, dim, interpolation = cv2.INTER_AREA) #redimensionamento da imagem 1
cv2.imwrite("image1RD.jpg", resized_image1)
resized_image2 = cv2.resize(image2, dim, interpolation = cv2.INTER_AREA) #redimensionamento da imagem 2
cv2.imwrite("image2RD.png", resized_image2)
dist = cv2.addWeighted(resized_image1, 0.5, resized_image2, 0.5, 0.0) #misturando as imagens
#temos cv2.addWeighted como o parametro para a trasnsição das imagens sendo sua equeção: dst =  a . img1 + b . img 2 + y
#sendo a = peso da imagem 1 ;  b= o peso da imagem 2 ; y = gama = 0
#cv2.imshow("Imagens Misturadas", dist)
plt.figure(figsize=(20, 10))
plt.imshow(dist)
```

```python
#questao 10 - operações lógicas AND e NOT
import cv2
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np

#Leitura das imagens 
image1 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Images/img1.jpg')
image2 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Images/img2.png')

# conversão para escala cinza
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

#redimensionado as imagens para o memso tamanho
dim = (1200, 800)
resized_gray1 = cv2.resize(gray1, dim, interpolation = cv2.INTER_AREA) #redimensionamento da imagem 1
resized_gray2 = cv2.resize(gray2, dim, interpolation = cv2.INTER_AREA) #redimensionamento da imagem 2

#AND
#equação da operação lógica AND para as duas imagens
dest_and = cv2.bitwise_and(resized_gray1, resized_gray2, mask=None)

#NOT
#equação da operação lógica NOT para cada imagem
dest_not1 = cv2.bitwise_not(resized_gray1, mask = None)
dest_not2 = cv2.bitwise_not(resized_gray2, mask = None)

plt.figure(figsize=(20, 10)) #tamanho das imagens a serem plotadas
plt.subplot(1,3,1), plt.imshow(dest_and), plt.title("Resultado AND") # plot da operação AND
plt.subplot(1,3,2), plt.imshow(dest_not1), plt.title("Resultado Not 1") # plot da operação NOT na imagem 1
plt.subplot(1,3,3), plt.imshow(dest_not2), plt.title("Resultado Not 2") # plot da operação NOT na imagem 2
```

```python
#questao 11 - escala e rotação de imagens sem interpolação

import cv2
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np

#rotação sem interpolação

image1 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Images/img1.jpg')
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
(alt, lar) = gray1.shape[:2] #captura altura e largura
centro = (lar // 2, alt // 2) #acha o centro da imagem
M = cv2.getRotationMatrix2D(centro, 45, 1.0) #centro-angulo-escala
gray1_rotacionada = cv2.warpAffine(gray1, M, (lar, alt))
plt.imshow(gray1_rotacionada)

#Escala sem interpolação
gray1_redimensionada = gray1[::2 ,::2] #definição do redimensionamento

plt.subplot(1,3,1), plt.imshow(gray1), plt.title("imagem original") # plot da imagem original

plt.figure(figsize=(15, 10)) #tamanho das imagens a serem plotadas
plt.subplot(1,3,2), plt.imshow(gray1_rotacionada), plt.title("Rotação 45 graus") #plot da imagem de rotação em 45º

plt.subplot(1,3,3), plt.imshow(gray1_redimensionada), plt.title("Ecala menor") #plot da imagem - diminuição de escala
```
