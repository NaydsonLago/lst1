```python
#questao 9
import cv2
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np

img_a = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Images/Fig0232(a)(partial_body_scan).tif", 0)
img_b = 255 - img_a

img_dst = np.zeros(img_a.shape[:2], dtype=np.float)
height, width = img_a.shape[:2]
z = 3 * np.sum(img_a) / (img_a.size)
for i in range(height):
    for j in range(width):
        img_dst[i, j] = max(img_a[i, j], z)

plt.figure(figsize=(15, 10))
plt.subplot(1,3,1), plt.imshow(img_a, 'gray'), plt.title("Image A"), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,2), plt.imshow(img_b, 'gray'), plt.title("Image B"), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,3), plt.imshow(img_dst, 'gray'), plt.title("Calibration"), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

```python
#questao 10
import cv2
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np

image1 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Images/img1.jpg')
image2 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Images/img2.png')
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
dim = (1200, 800)
resized_gray1 = cv2.resize(gray1, dim, interpolation = cv2.INTER_AREA) #redimensionamento da imagem 1
resized_gray2 = cv2.resize(gray2, dim, interpolation = cv2.INTER_AREA) #redimensionamento da imagem 2


#AND
#res = cv2.bitwise_and(image1,image2, mask= None)
dest_and = cv2.bitwise_and(resized_gray1, resized_gray2, mask=None)

#NOT
dest_not1 = cv2.bitwise_not(resized_gray1, mask = None)
dest_not2 = cv2.bitwise_not(resized_gray2, mask = None)
#cv2.imshow('Resultado da imagem 1 com NOT ', dest_not1)
#cv2.imshow('Resultado da imagem 2 com NOT', dest_not2)

plt.subplot(1,3,1), plt.imshow(dest_and), plt.title("Resultado AND")
plt.subplot(1,3,2), plt.imshow(dest_not1), plt.title("Resultado Not 1")
plt.subplot(1,3,3), plt.imshow(dest_not2), plt.title("Resultado Not 2")
```

```python
#questao 11

import cv2
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np

#rotaÃ§Ã£o sem interpolaÃ§Ã£o

image1 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Images/img1.jpg')
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
(alt, lar) = gray1.shape[:2] #captura altura e largura
centro = (lar // 2, alt // 2) #acha o centro
M = cv2.getRotationMatrix2D(centro, 45, 1.0) #centro-angulo-escala
gray1_rotacionada = cv2.warpAffine(gray1, M, (lar, alt))
plt.imshow(gray1_rotacionada)

#Escala sem interpolaÃ§Ã£o
gray1_redimensionada = gray1[::2 ,::2] #definiÃ§Ã£o do redimensionamento

plt.subplot(1,3,1), plt.imshow(gray1), plt.title("imagem original")
plt.subplot(1,3,2), plt.imshow(gray1_rotacionada), plt.title("RotaÃ§Ã£o 45 graus")
plt.subplot(1,3,3), plt.imshow(gray1_redimensionada), plt.title("Ecala menor")
```
